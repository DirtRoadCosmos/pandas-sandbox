```
matches = pd.read_csv("https://raw.githubusercontent.com/andytilia/pandas-sandbox/main/england-premier-league-matches-2018-to-2019-stats.csv")
matches.head()
```

```
ref_matches = pd.DataFrame(matches.groupby("referee")["timestamp"].count())
ref_matches
```

```
ref_hy = pd.DataFrame(matches.groupby("referee")["home_team_yellow_cards"].sum())
ref_ay = pd.DataFrame(matches.groupby("referee")["away_team_yellow_cards"].sum())
ref_hr = pd.DataFrame(matches.groupby("referee")["home_team_red_cards"].sum())
ref_ar = pd.DataFrame(matches.groupby("referee")["away_team_red_cards"].sum())
yellows = pd.merge(ref_hy, ref_ay, on='referee', how='inner')
reds = pd.merge(ref_hr, ref_ar, on='referee', how='inner')
ref_cards = pd.merge(yellows, reds, on="referee", how="inner")
ref_cards = pd.merge(ref_cards, ref_matches, on="referee", how="inner")
ref_cards.columns = ["hy", "ay", "hr", "ar", "m"]
ref_cards
```

```
ref_cards['yellows'] = ref_cards['ay']+ref_cards['hy']
ref_cards['ypg'] = ref_cards['yellows']/ref_cards['m']
ref_cards['reds'] = ref_cards['ar']+ref_cards['hr']
ref_cards['rpg'] = ref_cards['reds']/ref_cards['m']
ref_cards
```
